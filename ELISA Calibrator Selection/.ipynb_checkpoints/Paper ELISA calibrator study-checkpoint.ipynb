### Normalisation strategy
We have created standard curves using a variety of calibrators. columns are:

pg/ml = calibrator concentration
abs = absorbance value
calibrator = purified protein
test = the test number for each repeat
plate = the experiment plate number that was used; this is significant because plates 1-2 are commercial capture plates and plate 3 is an in house capture plate. all have the same antibody coating.

Most calibrators are tested in only once and in only one plate.
However, common to all three plastes is the calibrator "IBL-Uman"

We'll fit 4PL curves to each IBL-Uman test result
Then we'll interpolate the ABS at each given pg/ml.unique() listed in the data table
from there we can calculate the conversion value for each other calibrator at their given concentration.

### A brief detour  
To calculate the Signal-to-Noise ratio at 2000 pg/ml for the IBL kit with RnD and IBL calibrators

```python
def predict_4pl_abs_logx(c, A, B, C, D):
    """
    Predict y from log10 concentration using:
    y = A + (D - A) / (1 + exp(-B * (log10(c) - C)))
    """
    try:
        if c <= 0:
            return np.nan
        x = np.log10(c)
        return A + (D - A) / (1 + np.exp(-B * (x - C)))
    except Exception:
        return np.nan
```

## Calculate conversion factors  
We'll need to do this for all pg/ml values based on interpolated absorbance values from the 4PL curves generated above


```python
# Extract the full IBL-Uman set, minus the pg/ml == 0
IBL_all = elisa_data[(elisa_data['calibrator'] == 'IBL-Uman') & (elisa_data['pg/ml'] != 0)]
IBL_all
```

```python
def predict_4pl_abs_logx(c, A, B, C, D):
    """
    Predict y from log10 concentration using:
    y = A + (D - A) / (1 + exp(-B * (log10(c) - C)))
    """
    try:
        if c <= 0:
            return np.nan
        x = np.log10(c)
        return A + (D - A) / (1 + np.exp(-B * (x - C)))
    except Exception:
        return np.nan

    
# # Example usage
# A = IBL_1_metrics_df.loc[IBL_1_metrics_df['Metric'] == 'A (upper asymptote)', 'Value'].values[0]
# B = IBL_1_metrics_df.loc[IBL_1_metrics_df['Metric'] == 'B (slope)', 'Value'].values[0]
# C = IBL_1_metrics_df.loc[IBL_1_metrics_df['Metric'] == 'C (EC50)', 'Value'].values[0]
# D = IBL_1_metrics_df.loc[IBL_1_metrics_df['Metric'] == 'D (lower asymptote)', 'Value'].values[0]
# print(A,B,C,D)

# predict_4pl_abs_your_form(c=10000, A=A, B=B, C=C, D=D)
```

```python
# Create concentration list for all tested concentraitons
conc_list = IBL_all['pg/ml'].unique().tolist()
conc_list.sort(reverse=True)

# Calculate interpolated absorbance values
absorbance_dict = {}
for key, df in PL_metrics.items():
    A = df.loc[df['Metric'] == 'A (upper asymptote)', 'Value'].values[0]
    B = df.loc[df['Metric'] == 'B (slope)', 'Value'].values[0]
    C = df.loc[df['Metric'] == 'C (EC50)', 'Value'].values[0]
    D = df.loc[df['Metric'] == 'D (lower asymptote)', 'Value'].values[0]
    print(key, A, B, C, D)
    
    print('predicted ABS near LOD (30): ', predict_4pl_abs_logx(30, A = A, B = B, C = C, D = D))
    
    absorbance_list = []
    for conc in conc_list:
        absorbance = predict_4pl_abs_logx(conc, A = A, B = B, C = C, D = D)
        absorbance_list.append(absorbance)
    
    absorbance_dict[key] = absorbance_list
```

```python
# Create dataframe
IBL_Uman_4PL_df = pd.DataFrame(data=absorbance_dict, index=conc_list)

# Calculate conversion factors for each concentration point
IBL_Uman_4PL_df['plate 1 cf'] = 1
IBL_Uman_4PL_df['plate 2 cf'] = IBL_Uman_4PL_df['plate 1']/IBL_Uman_4PL_df['plate 2']
IBL_Uman_4PL_df['plate 3 cf'] = IBL_Uman_4PL_df['plate 1']/IBL_Uman_4PL_df['plate 3']

IBL_Uman_4PL_df
```

## Apply Conversion Factors

```python
# exclude pg/ml == 0 from the conversion results- they can't be mapped onto a log10 conc plot anyway. 
elisa_data_sub0 = elisa_data[elisa_data['pg/ml'] != 0]
elisa_data_sub0
```

```python
def get_abs_cf(row):
    pg = row['pg/ml']
    test_num = int(row['plate'])  # e.g., 1, 2, 3
    column_name = f'plate {test_num} cf'
    
    if pg in IBL_Uman_4PL_df.index and column_name in IBL_Uman_4PL_df.columns:
        return IBL_Uman_4PL_df.loc[pg, column_name]
    else:
        return np.nan
```

```python
elisa_data_sub0 = elisa_data_sub0.copy()
elisa_data_sub0['abs_cf'] = elisa_data_sub0.apply(get_abs_cf, axis=1)
elisa_data_sub0['abs_norm'] = elisa_data_sub0['abs'] * elisa_data_sub0['abs_cf']
elisa_data_sub0
```

## Plot elisa_data_sub0  

```python
# We'll drop the IBL-Uman values from plates 2 and 3
elisa_data_sub0_f = elisa_data_sub0[
    ~((elisa_data_sub0['calibrator'] == 'IBL-Uman') & 
      (elisa_data_sub0['plate'].isin([2, 3])))]

# We'll also drop where pg/ml >5000
elisa_data_sub0_ff =elisa_data_sub0_f[~(elisa_data_sub0['pg/ml'] >5000)]

# We'll also drop progen from plate 2. we'll keep the results from plate 3.
elisa_data_sub0_fff = elisa_data_sub0_ff[
    ~((elisa_data_sub0_ff['calibrator'] == 'Progen') & (elisa_data_sub0_ff['plate'] == 2))]

# Create log10(pg/ml)
elisa_data_sub0_fff = elisa_data_sub0_fff.copy()
elisa_data_sub0_fff['log10(pg/ml)'] = np.log10(elisa_data_sub0_fff['pg/ml'])

# inspect
elisa_data_sub0_fff
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import numpy as np



def xlog10_4PL(x, A, B, C, D):
    return A + (D - A) / (1 + np.exp(-B * (x - C)))

plt.figure(figsize=(10, 6))

for calibrator in elisa_data_sub0_fff['calibrator'].unique():
    df = elisa_data_sub0_fff[elisa_data_sub0_fff['calibrator'] == calibrator]
    x = df['log10(pg/ml)'].values
    y = df['abs_norm'].values

    if len(x) < 4:
        continue

    try:
        metrics_df, equation = robust_4pl_fit(x, y, transform_x="log10(x)", plot=False)

        popt = [
            metrics_df.loc[metrics_df['Metric'] == 'A (upper asymptote)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'B (slope)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'C (EC50)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'D (lower asymptote)', 'Value'].values[0],
        ]

        # Plot raw points without legend
        plt.scatter(10**x, y, alpha=0.4, s=20)

        # Plot fit line with label once
        x_smooth_log = np.linspace(min(x), max(x), 300)
        x_smooth = 10**x_smooth_log
        y_smooth = xlog10_4PL(x_smooth_log, *popt)
        plt.plot(x_smooth, y_smooth, label=calibrator)

    except Exception as e:
        print(f"Fit failed for {calibrator}: {e}")

plt.xscale('log')
plt.xlabel('Concentration (pg/ml)')
plt.ylabel('Absorbance (normalized)')
plt.title('Uman-type Ab NfL ELISA vs Calibrator type')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.gca().xaxis.set_major_formatter(ScalarFormatter())
plt.legend()
plt.tight_layout()
# plt.savefig("Uman-Type ELISA vs Calibrator Lin-Log.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import numpy as np
from matplotlib.lines import Line2D

def xlog10_4PL(x, A, B, C, D):
    return A + (D - A) / (1 + np.exp(-B * (x - C)))

# Set grayscale mode
grayscale = True

# Marker styles for distinct calibrators
marker_styles = ['o', 's', '^', 'D', 'v', 'p', '*', 'X', '<', '>']
legend_handles = []

plt.figure(figsize=(10, 6))

for idx, calibrator in enumerate(elisa_data_sub0_fff['calibrator'].unique()):
    df = elisa_data_sub0_fff[elisa_data_sub0_fff['calibrator'] == calibrator]
    x = df['log10(pg/ml)'].values
    y = df['abs_norm'].values

    if len(x) < 4:
        continue

    try:
        # 4PL fit
        metrics_df, equation = robust_4pl_fit(x, y, transform_x="log10(x)", plot=False)

        popt = [
            metrics_df.loc[metrics_df['Metric'] == 'A (upper asymptote)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'B (slope)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'C (EC50)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'D (lower asymptote)', 'Value'].values[0],
        ]

        # Style by calibrator
        marker = marker_styles[idx % len(marker_styles)]

        # Plot raw points
        plt.scatter(
            10**x, y,
            s=40,
            alpha=1.0,
            facecolors='none',
            edgecolors='black',
            linewidths=0.5,
            marker=marker
        )

        # Plot 4PL fit
        x_smooth_log = np.linspace(min(x), max(x), 300)
        x_smooth = 10**x_smooth_log
        y_smooth = xlog10_4PL(x_smooth_log, *popt)

        line, = plt.plot(x_smooth, y_smooth, color='black', linewidth=1.0)
        line.set_dashes([5, 2, 1, 2])  # dashed-dot line

        # Legend handle with marker only
        handle = Line2D([0], [0],
                        marker=marker,
                        linestyle='none',
                        color='none',
                        label=calibrator,
                        markerfacecolor='none',
                        markeredgecolor='black',
                        markeredgewidth=0.5,
                        markersize=6)
        legend_handles.append(handle)

    except Exception as e:
        print(f"Fit failed for {calibrator}: {e}")

# Axis settings
plt.xscale('log')
plt.xlabel('Concentration (pg/ml)')
plt.ylabel('Absorbance (normalized)')
plt.title('Uman-type Ab NfL ELISA vs Calibrator type')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.gca().xaxis.set_major_formatter(ScalarFormatter())

# Legend using marker shapes
plt.legend(handles=legend_handles, frameon=False, loc='best')

plt.tight_layout()
plt.savefig("Uman-Type ELISA vs Calibrator Lin-Log BW.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import numpy as np
from matplotlib.lines import Line2D

def xlog10_4PL(x, A, B, C, D):
    return A + (D - A) / (1 + np.exp(-B * (x - C)))

# Enable grayscale mode
grayscale = True

# Marker styles for visual distinction
marker_styles = ['o', 's', '^', 'D', 'v', 'p', '*', 'X', '<', '>']
legend_handles = []

plt.figure(figsize=(10, 6))

for idx, calibrator in enumerate(elisa_data_sub0_fff['calibrator'].unique()):
    df = elisa_data_sub0_fff[elisa_data_sub0_fff['calibrator'] == calibrator]
    x = df['log10(pg/ml)'].values
    y = df['abs_norm'].values

    if len(x) < 4 or np.any(y <= 0):
        continue  # Skip invalid entries for log-log scale

    try:
        # 4PL fit
        metrics_df, equation = robust_4pl_fit(x, y, transform_x="log10(x)", plot=False)

        popt = [
            metrics_df.loc[metrics_df['Metric'] == 'A (upper asymptote)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'B (slope)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'C (EC50)', 'Value'].values[0],
            metrics_df.loc[metrics_df['Metric'] == 'D (lower asymptote)', 'Value'].values[0],
        ]

        # Marker selection
        marker = marker_styles[idx % len(marker_styles)]

        # Plot raw data (hollow markers)
        plt.scatter(
            10**x, y,
            s=40,
            alpha=1.0,
            facecolors='none',
            edgecolors='black',
            linewidths=0.5,
            marker=marker
        )

        # Plot 4PL fit
        x_smooth_log = np.linspace(min(x), max(x), 300)
        x_smooth = 10**x_smooth_log
        y_smooth = xlog10_4PL(x_smooth_log, *popt)

        line, = plt.plot(x_smooth, y_smooth, color='black', linewidth=1.0)
        line.set_dashes([5, 2, 1, 2])  # dashed-dot

        # Legend handle (marker only)
        handle = Line2D([0], [0],
                        marker=marker,
                        linestyle='none',
                        color='none',
                        label=calibrator,
                        markerfacecolor='none',
                        markeredgecolor='black',
                        markeredgewidth=0.5,
                        markersize=6)
        legend_handles.append(handle)

    except Exception as e:
        print(f"Fit failed for {calibrator}: {e}")

# Log-log scaling
plt.xscale('log')
plt.yscale('log')

# Labels and ticks
plt.xlabel('Concentration (pg/ml)')
plt.ylabel('Absorbance (normalized, log10 scale)')
plt.title('Uman-type Ab NfL ELISA vs Calibrator type (log-log scale)')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.gca().xaxis.set_major_formatter(ScalarFormatter())
plt.gca().yaxis.set_major_formatter(ScalarFormatter())

# Legend (marker-based)
plt.legend(handles=legend_handles, frameon=False, loc='best')

plt.tight_layout()
plt.savefig("Uman-Type ELISA vs Calibrator Log-Log BW.png", dpi=300, bbox_inches='tight')
plt.show()
```
