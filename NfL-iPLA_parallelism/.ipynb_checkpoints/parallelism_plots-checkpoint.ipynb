## Load source data and reconstruct `lin_list_eff_trimmed`

```python
source_path = Path('parallelism_source_data.csv')
combined_df = pd.read_csv(source_path)
print(f'Loaded: {combined_df.shape[0]} rows x {combined_df.shape[1]} columns')
```

```python
lin_list = analyse_sample_linearity(combined_df)
```

```python
sample_map = defaultdict(list)
for i, df in enumerate(lin_list):
    for sid in df['sample_id'].unique():
        sample_map[sid].append(i)

lin_list_eff = []
for sid, indices in sample_map.items():
    if len(indices) == 1:
        lin_list_eff.append(lin_list[indices[0]])
    else:
        best_eff, best_idx = None, None
        for idx in indices:
            df = lin_list[idx]
            eff_vals = df[df['sample_id'] == sid]['efficiency'].unique()
            eff_closest = eff_vals[np.argmin(np.abs(eff_vals - 1))]
            if best_eff is None or abs(eff_closest - 1) < abs(best_eff - 1):
                best_eff = eff_closest
                best_idx = idx
        lin_list_eff.append(lin_list[best_idx][lin_list[best_idx]['sample_id'] == sid])
len(lin_list_eff)
```

```python
lin_list_eff_collapsed = []
for i, df in enumerate(lin_list_eff):
    numeric_cols = df.select_dtypes(include='number').columns.tolist()
    collapsed_rows = []
    for dilution_val, sub_df in df.groupby('dilution'):
        if len(sub_df) == 1:
            collapsed_rows.append(sub_df.iloc[0])
        else:
            differing_cols = []
            mean_row = sub_df.iloc[0].copy()
            for col in numeric_cols:
                if len(sub_df[col].dropna().unique()) > 1:
                    differing_cols.append(col)
                mean_row[col] = sub_df[col].mean()
            collapsed_rows.append(mean_row)
            if differing_cols:
                print(f'lin_list_eff[{i}] dilution {dilution_val}: mean for {differing_cols}')
    lin_list_eff_collapsed.append(pd.DataFrame(collapsed_rows).reset_index(drop=True))
```

```python
exclude_ids = [
    'NFL009LH', 'NFL001LH', 'NFL-QC-001', 'CSF Spike in IMD RMH 03/24',
    'NFL005LH[169.8]', 'NFL-QC-002', 'NFL-QC-M[19]',
    'IMD EQC2', 'NFL-QC-076'
]
exclude_substrings = []

lin_list_eff_trimmed = []
for df in lin_list_eff_collapsed:
    sid = df['sample_id'].unique()[0]
    if sid in exclude_ids:
        continue
    if any(sub in sid for sub in exclude_substrings):
        continue
    lin_list_eff_trimmed.append(df)

lin_list_eff_trimmed[8] = lin_list_eff_trimmed[8][lin_list_eff_trimmed[8]['dilution'] <= 3200]
print(f'lin_list_eff_trimmed: {len(lin_list_eff_trimmed)} samples')
```

## PUBLICATION PLOT 1 — Concentration vs Dilution (log₂–log₂)

```python
### PUBLICATION PLOT 1

from matplotlib.lines import Line2D

sns.set_context('notebook')
sns.set_style('whitegrid')

fig, ax = plt.subplots(figsize=(8, 6))

all_dilutions = np.concatenate([df['dilution'].values for df in lin_list_eff_trimmed])
all_concs = np.concatenate([df['SLR; log10(x); exc_std0; ct; mean_ng/L'].values for df in lin_list_eff_trimmed])

dil_tickvals  = np.arange(np.floor(np.log2(all_dilutions.min())), np.ceil(np.log2(all_dilutions.max())) + 1)
conc_tickvals = np.arange(np.floor(np.log2(all_concs.min())), np.ceil(np.log2(all_concs.max())) + 1)
dil_ticktext  = [str(int(2**v)) for v in dil_tickvals]
conc_ticktext = [str(round(2**v, 2)) for v in conc_tickvals]

for df in lin_list_eff_trimmed:
    sample_id = df['sample_id'].unique()[0]
    x = np.log2(df['dilution'])
    y = np.log2(df['SLR; log10(x); exc_std0; ct; mean_ng/L'])
    color = 'tomato' if 'CSF' in sample_id else 'royalblue'
    ax.plot(x, y, linewidth=1, color=color, alpha=0.4)
    ax.scatter(x, y, color=color, edgecolor='black', linewidth=0.5, s=15, zorder=3)

dilution_line_x = np.log2(10)
ax.axvline(x=dilution_line_x, color='black', linestyle='--', linewidth=1)
y_min, y_max = ax.get_ylim()
ax.text(dilution_line_x + 0.1, y_min + 0.01*(y_max - y_min),
        'minimal required dilution', color='black', fontsize=9, ha='left', va='bottom')

ax.set_xlabel('Dilution', fontsize=12)
ax.set_ylabel('Input Concentration (pg/L)', fontsize=12)
ax.set_xticks(dil_tickvals)
ax.set_xticklabels(dil_ticktext)
ax.set_yticks(conc_tickvals)
ax.set_yticklabels(conc_ticktext)
ax.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)
ax.set_axisbelow(True)

legend_elements = [
    Line2D([0],[0], color='red', lw=0.5, marker='o', markerfacecolor='tomato',
           markeredgecolor='black', label='CSF'),
    Line2D([0],[0], color='blue', lw=0.5, marker='o', markerfacecolor='royalblue',
           markeredgecolor='black', label='Plasma')
]
ax.legend(handles=legend_elements, title='Sample Type', loc='lower right', frameon=False)
sns.despine()
plt.tight_layout()
# plt.savefig('parallelism_plot1.svg', dpi=300, bbox_inches='tight')
plt.show()
```

## Build `matched_df_std_f` for PUBLICATION PLOT 2

Retained verbatim — standard curve rows are in the source CSV.

```python
summary_rows = []
for df in lin_list_eff_trimmed:
    summary_rows.append({
        'filename_csv': df['filename_csv'].iloc[0],
        'SLR; log10(x); exc_std0; ct; grad': df['SLR; log10(x); exc_std0; ct; grad'].iloc[0]
    })
std_grad_df = pd.DataFrame(summary_rows).drop_duplicates(subset='filename_csv', keep='first')
```

```python
matched_rows = []
for _, row in std_grad_df.iterrows():
    filename = row['filename_csv']
    grad = row['SLR; log10(x); exc_std0; ct; grad']
    match = combined_df[
        (combined_df['filename_csv'] == filename) &
        (combined_df['SLR; log10(x); exc_std0; ct; grad'] == grad)
    ][['filename_csv','sample_id','rep_id','ct; mean',
       'SLR; log10(x); exc_std0; ct; grad',
       'SLR; log10(x); exc_std0; ct; mean_ng/L',
       'SLR; log10(x); exc_std0; ct; PCReff']]
    matched_rows.append(match)

matched_df = pd.concat(matched_rows, ignore_index=True).drop_duplicates()
matched_df_std = matched_df[matched_df['sample_id'].str.contains('std', case=False, na=False)].copy()
```

```python
matched_df_std['log10_mean_ng/L'] = np.log10(matched_df_std['SLR; log10(x); exc_std0; ct; mean_ng/L'])

regression_rows = []
for name, group in matched_df_std.groupby('filename_csv'):
    x = group['log10_mean_ng/L'].values
    y = group['ct; mean'].values
    if len(x) < 2:
        continue
    slope, intercept, *_ = linregress(x, y)
    ct_targets = [12, 22, 32]
    predicted_x = [(ct - intercept) / slope for ct in ct_targets]
    regression_rows.append({
        'filename_csv': name, 'slope': slope, 'intercept': intercept,
        'log10_ct12': predicted_x[0], 'log10_ct22': predicted_x[1], 'log10_ct32': predicted_x[2]
    })

regression_df = pd.DataFrame(regression_rows)
matched_df_std = matched_df_std.merge(regression_df, on='filename_csv', how='left')

matched_df_std['ct12']       = matched_df_std['log10_ct12'] - 2.401413
matched_df_std['ct32']       = matched_df_std['log10_ct32'] + 0.567511
matched_df_std['ct22(lower)']= matched_df_std['log10_ct22'] - 2.401413
matched_df_std['ct22(upper)']= matched_df_std['log10_ct22'] + 0.567511

matched_df_std_f = matched_df_std[[
    'filename_csv', 'SLR; log10(x); exc_std0; ct; grad',
    'SLR; log10(x); exc_std0; ct; PCReff',
    'ct12', 'ct32', 'ct22(lower)', 'ct22(upper)'
]].drop_duplicates()
matched_df_std_f
```

## PUBLICATION PLOT 2 — Cq vs Dilution (Parallelism to Standard Curve)

```python
### PUBLICATION PLOT 2

from matplotlib.ticker import ScalarFormatter, MultipleLocator

fig, ax = plt.subplots(figsize=(8, 6))

# Sample linear fits
for df in lin_list_eff_trimmed:
    dil_vals = df['dilution'].values
    x_fit = np.logspace(np.log10(dil_vals.min()), np.log10(dil_vals.max()), 200)
    log10_inv_dil = np.log10(100 / x_fit)
    m = df['m'].iloc[0]
    c = df['c'].iloc[0]
    y_fit = m * log10_inv_dil + c
    ax.plot(x_fit, y_fit, color='royalblue', linewidth=0.7, zorder=6)

# Standard curve dashed segments
for _, row in matched_df_std_f.iterrows():
    try:
        d1 = 100 / (10**row['ct12'])
        d2 = 100 / (10**row['ct22(lower)'])
        d3 = 100 / (10**row['ct22(upper)'])
        d4 = 100 / (10**row['ct32'])
        ax.plot([d1, d2], [12, 22], color='tomato', linewidth=1, zorder=6)
        ax.plot([d3, d4], [22, 32], color='tomato', linewidth=1, zorder=6)
    except KeyError:
        pass

# Reference slope band
m_ref = -3.32
x_left, x_right = 4, 1400
x_ref = np.logspace(np.log10(x_left), np.log10(x_right), 300)
c_upper = 26     - m_ref * np.log10(100 / x_left)
c_lower = 10.8   - m_ref * np.log10(100 / x_left)
y_upper = m_ref * np.log10(100 / x_ref) + c_upper
y_lower = m_ref * np.log10(100 / x_ref) + c_lower
ax.fill_between(x_ref, y_lower, y_upper, color='lightgrey', alpha=0.35, zorder=2,
                label='Ideal theoretical reference slope (m = -3.32)')

# Axis formatting
ax.set_xscale('log')
ax.set_xlim(x_left, x_right)
ax.invert_xaxis()
custom_ticks = [5, 20, 80, 320, 1280]
ax.set_xticks(custom_ticks)
ax.get_xaxis().set_major_formatter(ScalarFormatter())
ax.set_xticklabels([str(t) for t in custom_ticks])
ax.yaxis.set_major_locator(MultipleLocator(2))
ax.grid(True, which='both', linestyle='--', color='black', alpha=1, zorder=5, linewidth=0.2)

legend_elements = [
    Line2D([0],[0], color='royalblue', lw=1, label='Samples'),
    Line2D([0],[0], color='tomato', lw=1, label='Standards'),
]
ax.legend(handles=legend_elements + ax.get_legend_handles_labels()[0],
          loc='lower left', frameon=True, framealpha=1)

ax.set_xlabel('Sample Dilution')
ax.set_ylabel('Cq (Mean)')
plt.tight_layout()
# plt.savefig('parallelism_plot2.svg', dpi=300, bbox_inches='tight')
plt.show()
```
