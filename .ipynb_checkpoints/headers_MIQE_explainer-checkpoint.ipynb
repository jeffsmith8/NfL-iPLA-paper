# `headers_MIQE.txt` — Purpose, Usage, and Column Definitions

## What is this file?

`headers_MIQE.txt` is a plain-text list of column headers — one per line — that defines the **MIQE-compliant subset** of the full ProxiPal output table.

The full output table produced by ProxiPal can exceed **6000 columns**, most of which represent statistical outputs from multiple curve-fitting models (SLR, 4PL, 5PL, etc.) across multiple quantification strategies. Only a small fraction of these are required for journal submission under the [MIQE guidelines](https://doi.org/10.1373/clinchem.2008.112797) (Bustin et al., *Clinical Chemistry*, 2009).

`headers_MIQE.txt` is the authoritative filter list used to trim ProxiPal tables down to publication-relevant columns.  
It lives in the parent directory alongside `ProxiPal.py`.  

---
## What is ProxiPal?
ProxiPal is a python package created to support PLA development, early stage validation studies, service support and assay kit production.  
  
[ProxiPal on GitHub](https://github.com/proxipal-pcr/proxipal)
  
The development of iPLA has been conducted across many conditions, variables, instruments, software, and mathematical models. ProxiPal consolidates all these outputs into .csv tables for record keeping, monitoring, and reporting. After data is extracted for analysis (from ProxiPal tables) we use headers_MIQE.txt to trim tables to the relevant data. In this manner, outputs from user guided excel calculation and metadata, OEM software, rdml outputs, and python statistics can be reported in a single .csv table.

_NB: Information may be incomplete for some instruments_

## Column definitions

Headers are grouped by category. MIQE classification is noted where applicable: **E** = Essential, **D** = Desirable.

---

### 1. Sample & run identification
Defined in user submitted excel experimental design templates

| Column | Description | MIQE |
|---|---|---|
| `well` | Physical well identifier on the plate (e.g. A1, B3) | — |
| `position` | Numeric well position index | — |
| `target` | Gene/assay target name | E |
| `sample_id` | Sample identifier | E |
| `tube_id` | Tube-level identifier (i.e. some sample_ids represent patients with multiple collection timepoints) | E |
| `rep_id` | Replicate identifier (technical or biological, as labelled) | E |
| `dilution` | Dilution factor applied to sample prior to qPCR | E |
| `calibrator` | Identifies which well/sample serves as the calibrator for relative quantification | E |

---
### 2. User-defined variables

Defined in user submitted excel experimental design templates (e.g. matrix type, collection site, treatment group).

| Column | Description | MIQE |
|---|---|---|
| `var_1` – `var_4` | User-defined categorical or continuous variable values | D |
| `var_1 (desc)` – `var_4 (desc)` | Plain-language description of what each `var_n` field represents | D |
| `usr_std` | Flags which std curve was used to calculate metrics for the given row | E |
| `usr_raw_ng/L` | Raw concentration estimate in ng/L for this well (pre-averaging) | E |
| `usr_mean_ng/L` | Mean concentration in ng/L across replicates | E |

---

### 3. Instrument metadata

Required by MIQE for assay reproducibility assessment and inter-laboratory comparison.

| Column | Description | MIQE |
|---|---|---|
| `InstrumentName` | Instrument model name (e.g. QuantStudio 7) | E |
| `InstrumentSerialNumber` | Instrument serial number | E |
| `InstrumentType` | Instrument platform type | E |
| `BlockType` | Thermal block type (e.g. 96-well, Fast) | D |
| `ExperimentRunStartTime` | Timestamp when the run started | D |
| `ExperimentRunStopTime` | Timestamp when the run ended | D |
| `ExperimentType` | Experiment mode (e.g. Standard Curve, Comparative Ct) | E |
| `DateCreated` | Date the experiment file was created | D |
| `Chemistry` | Fluorescence chemistry used (SYBR Green or TaqMan/probe-based) | E |
| `PassiveReference` | Passive reference dye used for well-to-well normalisation (typically ROX) | E |
| `QuantificationCycleMethod` | Algorithm used to determine Cq (e.g. auto threshold, fit-point) | E |

---

### 4. Primary qPCR instrument output
The core per-well quantitative values from the instrument's own OEM software.

| Column | Description | MIQE |
|---|---|---|
| `ct` | Quantification cycle (Cq) for this well, as reported by instrument | E |
| `ct; mean` | Mean Cq across replicates | E |
| `tm` | Melt temperature (°C) from melt curve analysis. SYBR Green assays only — used as primary specificity check | E (SYBR) |
| `threshold` | Fluorescence threshold value used for Cq determination | E |

---

### 12. Standard curve statistics — Ct-based (SLR)
Simple linear regression of Ct vs log₁₀(known concentration). Standard curve std0 (the zero standard / blank) is excluded from the fit. This is the conventional MIQE calibration curve.

| Column | Description | MIQE |
|---|---|---|
| `SLR; log10(x); exc_std0; ct; grad` | Slope of the calibration curve. Ideally −3.32 (100% efficiency). Used to calculate PCR efficiency. | E |
| `SLR; log10(x); exc_std0; ct; y_int` | Y-intercept of the calibration curve | E |
| `SLR; log10(x); exc_std0; ct; equation` | Full regression equation as a string | D |
| `SLR; log10(x); exc_std0; ct; R2` | R² of the calibration curve fit. MIQE requires R² ≥ 0.98 as minimum. | E |
| `SLR; log10(x); exc_std0; ct; PCReff` | PCR efficiency calculated from the slope: E = 10^(−1/slope) − 1. Should be 0.90–1.10. | E |
| `SLR; log10(x); exc_std0; ct; raw_ng/L` | Back-calculated concentration for each well from the Ct-based curve (ng/L) | E |
| `SLR; log10(x); exc_std0; ct; mean_ng/L` | Mean back-calculated concentration across replicates (ng/L) | E |
| `SLR; log10(x); exc_std0; ct; raw_recovery` | Per-replicate recovery (back-calculated / known × 100%) | E |
| `SLR; log10(x); exc_std0; ct; mean_recovery` | Mean recovery across replicates | E |

---

### 13. Standard curve statistics — RDML log₂N0-based (SLR)
Same SLR framework as above, but using efficiency-corrected log₂N0 as the response variable instead of raw Ct. 

| Column | Description | MIQE |
|---|---|---|
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; grad` | Slope of log₂N0 vs log₁₀(known conc) calibration curve | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; y_int` | Y-intercept | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; equation` | Full regression equation as string | D |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; R2` | R² of fit | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; PCReff` | PCR efficiency derived from this fit | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; raw_ng/L` | Back-calculated concentration per well (ng/L) | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; mean_ng/L` | Mean concentration across replicates (ng/L) | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; raw_recovery` | Per-replicate recovery | E |
| `SLR; log10(x); exc_std0; rdml_log2N0 (mean eff) - no plateau - stat efficiency; mean_recovery` | Mean recovery across replicates | E |

### 5. RDML sample and target metadata
Fields from the RDML (Real-Time PCR Data Markup Language) standard. 

| Column | Description | MIQE |
|---|---|---|
| `rdml_sample` | Sample name as recorded in RDML | E |
| `rdml_sample type` | Sample type classification (e.g. unkn, std, ntc, nac) | E |
| `rdml_sample nucleotide` | Nucleotide type of the sample (DNA, RNA, cDNA) | E |
| `rdml_target` | Target name as recorded in RDML | E |
| `rdml_target chemistry` | Detection chemistry for this target (hydrolysis probe, SYBR, etc.) | E |
| `rdml_excluded` | Boolean flag — whether RDML has marked this well as excluded from analysis | E |
| `rdml_note` | Free-text note attached to this well in RDML | D |

---

### 6. RDML baseline and threshold
Parameters governing how the amplification signal is baseline-corrected and thresholded.

| Column | Description | MIQE |
|---|---|---|
| `rdml_baseline` | Baseline correction method applied | E |
| `rdml_lower limit` | Lower cycle boundary of the baseline window | E |
| `rdml_upper limit` | Upper cycle boundary of the baseline window | E |
| `rdml_common threshold` | Single threshold applied uniformly across all wells in the run | E |
| `rdml_group threshold` | Threshold applied to a defined subset/group of wells | E |

---

### 7. RDML log-linear phase parameters
These define the window of cycles used for efficiency calculation — the exponential amplification phase before plateau.

| Column | Description | MIQE |
|---|---|---|
| `rdml_n in log phase` | Number of cycles identified as being within the log-linear phase | E |
| `rdml_last log cycle` | Last cycle number of the log-linear phase | E |
| `rdml_n included` | Number of cycles actually included in the efficiency fit | E |
| `rdml_log lin cycle` | Cycle number at the midpoint of the log-linear phase | D |
| `rdml_log lin fluorescence` | Fluorescence value at the midpoint of the log-linear phase | D |

---

### 8. RDML PCR efficiency — primary values

MIQE requires PCR efficiency to be reported. The primary efficiency values used for quantification are the **no-plateau, stat-efficiency** variants, which use per-well efficiency correction and exclude plateau-affected cycles.

| Column | Description | MIQE |
|---|---|---|
| `rdml_indiv PCR eff` | Per-well (individual) PCR efficiency calculated from the log-linear phase fit | E |
| `rdml_R2` | R² of the per-well log-linear phase fit — quality of the efficiency estimate | E |
| `rdml_mean PCR eff - no plateau - stat efficiency` | Mean PCR efficiency across wells, excluding plateau cycles, using statistical efficiency estimation. **Primary efficiency value for reporting.** | E |
| `rdml_standard error of the stat PCR eff - no plateau - stat efficiency` | Standard error of the mean PCR efficiency. Satisfies MIQE requirement for confidence interval / SE on efficiency | E |

---

### 9. RDML quantification outputs — primary
Efficiency-corrected quantification values. 

| Column | Description | MIQE |
|---|---|---|
| `rdml_N0 (mean eff) - no plateau - stat efficiency` | Initial target quantity (N0) estimated using mean efficiency, no-plateau, stat-efficiency method. Absolute copy number estimate. | E |
| `rdml_Cq (mean eff) - no plateau - stat efficiency` | Efficiency-corrected Cq recalculated using mean efficiency. More comparable across wells and runs than raw Ct. | E |
| `rdml_log2N0 (mean eff) - no plateau - stat efficiency` | log₂-transformed N0. Used for relative quantification and fold-change calculations. | E |
| `rdml_log2N0 (mean eff) - no plateau - stat efficiency; mean` | Mean log₂N0 across replicates. | E |

---

### 10. RDML quality control flags
Boolean flags generated by RDML analysis. Essential for MIQE outlier reporting and transparency.

| Column | Description |
|---|---|
| `rdml_amplification` | Well showed detectable amplification |
| `rdml_baseline error` | Baseline correction failed or produced an error |
| `rdml_instable baseline` | Baseline was unstable (high variance in pre-amplification cycles) |
| `rdml_plateau` | Amplification reached a clear plateau — plateau cycles excluded from efficiency fit |
| `rdml_noisy sample` | High noise detected in the amplification curve |
| `rdml_PCR efficiency outside mean rage` | Per-well efficiency falls outside the acceptable range around the group mean |
| `rdml_PCR efficiency outside mean rage - no plateau` | Same, calculated excluding plateau cycles |
| `rdml_PCR efficiency outlier` | Per-well efficiency is a statistical outlier |
| `rdml_PCR efficiency outlier - no plateau` | Same, calculated excluding plateau cycles |
| `rdml_short log lin phase` | Log-linear phase is too short for a reliable efficiency estimate |
| `rdml_Cq is shifting` | Cq value shifts with threshold — indicates poor signal or inhibition |
| `rdml_too low Cq eff` | Cq is too low for reliable efficiency calculation |
| `rdml_too low Cq N0` | Cq is too low for reliable N0 estimation |
| `rdml_used for W-o-L setting` | Well was used to define the Window-of-Linearity for the run |

---

### 11. Known concentration inputs
Used for standard curve construction and recovery calculation.

| Column | Description | MIQE |
|---|---|---|
| `py_known_conc` | Known concentration of the standard curve point (linear scale, ng/L) | E |
| `py_known_conc_log10` | log₁₀-transformed known concentration. X-axis variable for SLR calibration curve fitting. | E |

### 14. Raw fluorescence data (EDS exports)
Per-cycle, per-well fluorescence data exported from the instrument EDS file.

#### Raw multiwell reads
| Column | Description |
|---|---|
| `well (eds; raw data)` | Well identifier |
| `well position (eds; raw data)` | Well position index |
| `cycle (eds; raw data)` | PCR cycle number |
| `x1-m1` – `x5-m5 (eds; raw data)` | Raw fluorescence reads per detector channel per cycle (pre-normalisation) |

#### Amplification curve data
| Column | Description |
|---|---|
| `well (eds; amplification data)` | Well identifier |
| `cycle (eds; amplification data)` | PCR cycle number |
| `target name (eds; amplification data)` | Target name for this well |
| `rn (eds; amplification data)` | Rn — reporter fluorescence normalised to passive reference (ROX). Raw signal used for Cq determination. |
| `delta rn (eds; amplification data)` | ΔRn — baseline-subtracted Rn. The amplification curve as displayed in instrument software. |

#### Multicomponent data (dye channel decomposition)
| Column | Description |
|---|---|
| `well (eds; multicomponent data)` | Well identifier |
| `well position (eds; multicomponent data)` | Well position index |
| `cycle (eds; multicomponent data)` | PCR cycle number |
| `rox (eds; multicomponent data)` | ROX passive reference signal per cycle |
| `sybr (eds; multicomponent data)` | SYBR Green channel signal per cycle |
| `fam (eds; multicomponent data)` | FAM channel signal per cycle |
| `cy5 (eds; multicomponent data)` | CY5 channel signal per cycle |

#### Melt curve raw data
| Column | Description |
|---|---|
| `well (eds; melt curve raw data)` | Well identifier |
| `well position (eds; melt curve raw data)` | Well position index |
| `reading (eds; melt curve raw data)` | Reading index within the melt ramp |
| `temperature (eds; melt curve raw data)` | Temperature (°C) at each melt curve reading |
| `fluorescence (eds; melt curve raw data)` | Fluorescence at each temperature step |
| `derivative (eds; melt curve raw data)` | −dF/dT derivative. Peak of this curve defines the reported Tm. |
