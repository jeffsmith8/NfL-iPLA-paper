```python
clone_data = pd.read_csv('SNR_all.csv')
clone_data
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the paired column plot
labels = clone_data['Label']
absorbance_values_ibl = clone_data['IBL_SNR']
absorbance_values_pk = clone_data['PK_SNR']

# Create the paired column plot
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

plt.figure(figsize=(14, 7))
plt.bar(x - width/2, absorbance_values_ibl, width, color='lightblue', label='IBL-NfL')
plt.bar(x + width/2, absorbance_values_pk, width, color='lightcoral', label='RnD-NfL')

# Add labels, title, and legend
plt.xticks(x, labels, rotation=45, ha='right')
plt.xlabel("Antibody Pairs", fontsize=12)
plt.ylabel("Signal-to-Noise Ratio (2000pg/ml)", fontsize=12)
plt.title("Signal-Noise Ratio of paired Ab clones using IBL-Uman bNfL or PK rhNfL", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which='major', axis='y', linestyle=':', alpha=0.7)  # Grid added here

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load your data
df = hytest_data

# Reshape for heatmap
df_melted = df.melt(id_vars=["Capture", "Detector"], var_name="Protein", value_name="SNR")
heatmap_data = df_melted.pivot_table(index=["Capture", "Detector"], columns="Protein", values="SNR")

# Sort by multi-index for consistent blocks
heatmap_data_sorted = heatmap_data.sort_index(level=["Capture", "Detector"])

# Begin plotting
fig, ax = plt.subplots(figsize=(14, 10))

# Draw the heatmap with hidden y-ticklabels
sns.heatmap(
    heatmap_data_sorted,
    cmap="viridis",
    ax=ax,
    cbar_kws={"label": "SNR"},
    linewidths=0.5,
    linecolor='gray',
    xticklabels=True,
    yticklabels=False  # we'll add custom y-axis text
)

# Get row info
y_positions = np.arange(len(heatmap_data_sorted))[::-1] + 0.5
detectors = heatmap_data_sorted.index.get_level_values("Detector")
captures = heatmap_data_sorted.index.get_level_values("Capture")

# Draw Detector labels for every row
for y, det in zip(y_positions, detectors):
    ax.text(-0.2, y, det, va='center', ha='right', fontsize=10)

# Draw Capture group labels once per block
prev_capture = None
for i, (y, cap) in enumerate(zip(y_positions, captures)):
    if cap != prev_capture:
        block_size = list(captures).count(cap)
        ax.text(-1.8, y - (block_size - 1) / 2, cap,
                va='center', ha='right', fontsize=12, fontweight='bold')
        prev_capture = cap

# Final annotations
ax.set_xlabel("Protein Source", fontsize=12)
ax.set_ylabel("Detector Antibody", fontsize=12)
ax.set_title("SNR Heatmap: Detectors by Protein Source, Grouped by Capture Antibody", fontsize=14)
plt.tight_layout()
plt.show()
```
